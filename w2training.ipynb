```python
def display_list(li,i):
  if i==len(li):
    return 'done'
  print(li[i])
  return display_list(li,i+1)
print(display_list([10,20,30,40,50,60],0))
```

```python
def mult(n):
  for i in range(1,11):
    print(n,'X',i,'=',n*i)
mult(19)
```

```python
def mult(n,i):
  if i<11:
    print(n,'X',i,'=',n*i)
    return mult(n,i+1)
  elif i==11:
    return 'Done'
  else:
    return 'input error'

mult(19,1)
```
